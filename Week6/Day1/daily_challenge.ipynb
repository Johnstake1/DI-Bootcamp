```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

cities = ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix', 'Philadelphia', 'San Antonio', 'San Diego', 'Dallas', 'San Jose']
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

temperature = np.random.uniform(-5,35, (10,12))

df=pd.DataFrame(temperature, index=cities, columns=months)
df
```

```python
#Calculate the annual average temperature for each city
annual_avg = df.mean(axis=1)
annual_avg

#Identify the city with the highest and lowest average temperature for the year

highest_avg = annual_avg.idxmax()
print(f'The city with the highest average temperature is {highest_avg}')

lowest_avg = annual_avg.idxmin()
print(f'The city with the lowest average temperature is {lowest_avg}')
```

```python
# Plot the temperatures for each city across the months
plt.figure(figsize=(12, 8))
for city in cities:
    plt.plot(months, df.loc[city], label=city)

plt.xlabel('Months')
plt.ylabel('Temperature (°C)')
plt.title('Monthly Temperatures for Each City')
plt.legend(loc='upper right')
plt.grid(True)
plt.show()
```
