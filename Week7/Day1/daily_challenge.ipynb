```python
url2 = 'webpage.html'

soup = BeautifulSoup(open(url2), 'html.parser')
```

```python
title = soup.find('title')

print(title.text)
```

```python
p_tags = soup.find_all('p', class_="f5 color-fg-muted mb-0 mt-1")

for ptags in p_tags:
    print(ptags.text)

print(p_tags)
```

```python
for ptags in p_tags:
    pp_tags = ptags.text
    print(pp_tags)

scrape_dict[title] = {"Description": pp_tags}

print(scrape_dict)
```
