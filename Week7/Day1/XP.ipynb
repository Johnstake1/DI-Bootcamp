🌟 Exercise 1 : Parsing HTML with BeautifulSoup


```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(open('XP.html'), 'html.parser')

title = soup.find('title')

print(title.text)
```

```python
p_tags = soup.find_all('p')

for ptags in p_tags:
    print(ptags.text)
```

```python
links = soup.find_all('a')
for link in links:
    print(link['href']) 

```

🌟 Exercise 2 : Scraping robots.txt from Wikipedia


```python
import requests
from bs4 import BeautifulSoup

url = 'https://en.wikipedia.org/wiki/Robots.txt'

response = requests.get(url)

robot_txt = BeautifulSoup(response.text, 'html.parser')

print(robot_txt.prettify())
```

🌟 Exercise 3 : Extracting Headers from Wikipedia’s Main Page


```python
url = 'https://en.wikipedia.org'

response = requests.get(url)

wiki_main_page = BeautifulSoup(response.text, 'html.parser')

#Lets find all the header tags in the page.

header_tags = wiki_main_page.find_all(['h1', 'h2', 'h3', 'h4', 'h5', 'h6'])

for header in header_tags:
    print(header.text)
```

🌟 Exercise 4 : Checking for Page Title


```python

def get_title(url):
    response = requests.get(url)
    website = BeautifulSoup(response.text, 'html.parser')
    return website.title.text


url = 'https://stackoverflow.com/questions/33282368/plotting-a-2d-heatmap' #insert url

get_title(url)
```

🌟 Exercise 5 : Analyzing US-CERT Security Alerts


```python
def get_alert(url):
    response = requests.get(url)
    website = BeautifulSoup(response.text, 'html.parser')
    alert = website.find_all('div', class_='c-view__row')
    return print(f'The number of alerts is: {len(alert)}')

url = 'https://www.cisa.gov/news-events/cybersecurity-advisories?f%5B0%5D=advisory_type%3A93'

get_alert(url)
```
