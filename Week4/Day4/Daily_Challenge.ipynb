```python
import pandas as pd

salaries = pd.read_csv('/content/datascience_salaries.csv')

salaries.head()
```

```python
salaries.info()
```

```python
salaries.describe() #The data shows a big spread between percentiles, lets try to normalize it
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
salaries['salary_norm'] = scaler.fit_transform(salaries[['salary']])
```

```python
salaries.head()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=1)
salaries['salary_pca'] = pca.fit_transform(salaries[['salary']])
```

```python
salaries_exp_level = salaries.groupby('experience_level')['salary'].agg(['mean', 'median']).reset_index()
```

```python
salaries_exp_level.head() #conclusion: the mean>median. Therefore it is skewed to the right (positive skewness).
```
