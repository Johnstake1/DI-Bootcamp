```python
import pandas as pd

titanic_data = pd.read_csv('train.csv')
titanic_data.head()
```

```python
#Create a copy of the dataframe. From now on we will use titanic_data_frame

titanic_data_frame = titanic_data.copy()
```

```python
titanic_data_frame.info()
titanic_data_frame.head()
```

 Exercise 1: Duplicate Detection and Removal

```python
titanic_data_frame = titanic_data_frame.drop_duplicates(subset = None, keep='first', inplace=False) #removes duplicated rows
```

```python
titanic_data_frame.info() #verifying again number of rows looks good. No duplicates found!
```

 Exercise 2: Handling Missing Values

```python
titanic_data_frame.isnull().sum() #checking for missing values.

# The only important data that we need to fill would be age.
# Cabin is not so relevant
```

You may impute by doing a linear regression or by using another statistic approach such as mode, meadian or mean. We will choose to do a fill using the median values.

```python
titanic_data_frame.describe()
```

We will add a feature called filled age for comparison reasons.

```python
titanic_data_frame['filled_Age'] = titanic_data_frame['Age'].fillna(titanic_data_frame.groupby(['Pclass','Sex'])['Age'].transform('median'))
```

```python
titanic_data_frame.describe()
```

 Exercise 3: Feature Engineering

```python
titanic_data_frame.head()
```

```python
titanic_data_frame['Parch'].unique()
```

```python
titanic_data_frame['SibSp'].unique()
```

```python
titanic_data_frame['FamilySize'] = titanic_data_frame['Parch'] + titanic_data_frame['SibSp']
```

Encoding for EX6 FYI

```python
titanic_data_frame['Sex'].unique() #converting categorical values or qualitative values into quants
titanic_data_frame['Sex'] = titanic_data_frame['Sex'].map({'male':0, 'female':1}) #another method is to do a get_dummy
```

Exercise 4: Outlier Detection and Handling. We will use the IQR method

```python
titanic_data_frame.describe()
```

```python
 Q1 = titanic_data_frame[['Age', 'Fare']].quantile(0.25)
 Q3 = titanic_data_frame[['Age', 'Fare']].quantile(0.75)
 IQR = Q3 - Q1

titanic_data_frame = titanic_data_frame[~((titanic_data_frame[['Age', 'Fare']]< (Q1-1.5*IQR) | (titanic_data_frame[['Age', 'Fare']]> (Q3+ 1.5*IQR)))).any(axis=1)]
```
