Your Task : Create a comprehensive visual analysis of the car sales dataset using different plotting libraries.

```python
import pandas as pd
import matplotlib.pyplot as plt
from plotnine import *
import numpy as np
import seaborn as sns
import plotly.express as px


car_dataset = pd.read_csv('USA_cars_datasets.csv')
car_dataset.info()
```

```python
car_dataset.head()
```

```python
car_dataset.rename(columns={'Unnamed: 0': 'index'}, inplace=True)
car_dataset.head()
```

```python
car_dataset.info()
```

Create a histogram showing the distribution of car prices.



```python
sns.histplot(car_dataset['price'], bins=30, kde=True)
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Distribution of Car Prices')
plt.show()
```

Create a box plot showing price distribution by brand (top 10 brands).



```python
brands = car_dataset['brand'].value_counts().nlargest(10).index

top10_brands_data = car_dataset[car_dataset['brand'].isin(brands)]

plt.figure(figsize=(12, 6))
sns.boxplot(x='brand', y='price', data=top10_brands_data)
plt.xlabel('Brand')
plt.ylabel('Price')
plt.title('Price Distribution by Brand (Top 10 Brands)')
plt.xticks(rotation=45)
plt.show()
```
